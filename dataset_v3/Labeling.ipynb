## tcp +arp,icmp

```python
df1=pd.read_csv('csv/in-tcp-1.csv')
df2=pd.read_csv('csv/in-tcp-2.csv')
df3=pd.read_csv('csv/in-tcp-3.csv')
df4=pd.read_csv('csv/in-tcp-4.csv')
df5=pd.read_csv('csv/in-tcp-5.csv')
df6=pd.read_csv('csv/in-tcp-6.csv')
df7=pd.read_csv('csv/in-tcp-7.csv')

intcp = pd.concat([df1, df2, df3, df4, df5, df6, df7])

del df1, df2, df3, df4, df5, df6, df7
```

```python
# tcp && eth.addr == 00:0c:29:fa:80:88
TCP_mask = (intcp['Protocol Type'] == 6) & ((intcp['src_mac'] == '00:0c:29:fa:80:88') | (intcp['dst_mac'] == '00:0c:29:fa:80:88'))
# udp && eth.addr == 00:0c:29:fa:80:88
UDP_mask = (intcp['Protocol Type'] == 17) & ((intcp['src_mac'] == '00:0c:29:fa:80:88') | (intcp['dst_mac'] == '00:0c:29:fa:80:88'))
# icmp && eth.addr == 00:0c:29:fa:80:88
ICMP_mask = (intcp['Protocol Type'] == 1) & ((intcp['src_mac'] == '00:0c:29:fa:80:88') | (intcp['dst_mac'] == '00:0c:29:fa:80:88'))
# arp && eth.addr == 00:0c:29:fa:80:88
ARP_mask = (intcp['Protocol Type'] == 0) & ((intcp['src_mac'] == '00:0c:29:fa:80:88') | (intcp['dst_mac'] == '00:0c:29:fa:80:88'))
```

```python
TCP_mask.sum(), UDP_mask.sum(), ICMP_mask.sum(), ARP_mask.sum()
```

```python
intcp['Label']='benign'

intcp.loc[TCP_mask, 'Label'] = 'TCP'
intcp.loc[UDP_mask, 'Label'] = 'UDP'
intcp.loc[ICMP_mask, 'Label'] = 'ICMP'
intcp.loc[ARP_mask, 'Label'] = 'ARP'
```
