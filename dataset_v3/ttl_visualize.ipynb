```python
import pandas as pd
import numpy as np
from pandas import read_csv as rc
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Thiết lập phong cách cho biểu đồ
# sns.set(style="whitegrid")

def make_visualization(data, label = ''):
	# Tạo biểu đồ phân phối cho cột TTL
	plt.figure(figsize=(10, 6))
	sns.histplot(data, bins=30, kde=True, color='blue')

	# Thêm tiêu đề và nhãn
	plt.title(label + ' Distribution')
	plt.xlabel(label + ' Value')
	plt.ylabel('Frequency')

	# Hiển thị biểu đồ
	plt.show()
```

```python
def make_visualization_v2(list_label, list_of_data):
    for label, data in zip(list_label, list_of_data):
        print(label)
        make_visualization(data, label)
```

```python
df = rc('dataset_v3.csv')
```

```python
df.Label.value_counts()
```

```python
df.Duration.value_counts().to_numpy()
```

```python
du_benign = df[df.Label == 'benign'].Duration
du_TCP = df[df.Label == 'TCP'].Duration
du_UDP = df[df.Label == 'UDP'].Duration
du_ICMP = df[df.Label == 'ICMP'].Duration
du_ARP = df[df.Label == 'ARP'].Duration
```

```python
list_data = [du_benign, du_TCP, du_UDP, du_ICMP, du_ARP]

make_visualization_v2(['du_benign', 'du_TCP', 'du_UDP', 'du_ICMP', 'du_ARP'], list_data)
```

```python
# Tạo biểu đồ phân phối cho tất cả các loại trên cùng một biểu đồ
plt.figure(figsize=(14, 8))

sns.histplot(du_benign, bins=30, kde=True, color='blue', label='Benign', alpha=0.5)
sns.histplot(du_TCP, bins=30, kde=True, color='green', label='TCP', alpha=0.5)
sns.histplot(du_UDP, bins=30, kde=True, color='red', label='UDP', alpha=0.5)
sns.histplot(du_ICMP, bins=30, kde=True, color='purple', label='ICMP', alpha=0.5)
sns.histplot(du_ARP, bins=30, kde=True, color='orange', label='ARP', alpha=0.5)

# Thêm tiêu đề và nhãn
plt.title('Distribution of Duration for Different Labels')
plt.xlabel('Duration Value')
plt.ylabel('Frequency')
plt.legend()

# Hiển thị biểu đồ
plt.show()
```
